# Python program that generate a part of the Piccadilly line using graph data structure

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph object
G = nx.Graph()

# Add nodes (stations) with positions
stations = {
    'Hyde Park Corner': (0, 0),
    'Green Park': (0.6, 0.6),
    'Piccadilly Circus': (1.6, 0.6),
    'Leicester Square': (2.8, 0.6),
    'Covent Garden': (3, 1),
    'Holborn': (3.2, 1.4)
}

for station, pos in stations.items():
    G.add_node(station, pos=pos)

# Add edges (connections between stations) with distances
edges = [
    ('Hyde Park Corner', 'Green Park', 0.8),  # Distance in kilometers
    ('Green Park', 'Piccadilly Circus', 0.8),
    ('Piccadilly Circus', 'Leicester Square', 1),
    ('Leicester Square', 'Covent Garden', 0.4),
    ('Covent Garden', 'Holborn', 0.4)
]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Get positions of nodes
pos = nx.get_node_attributes(G, 'pos')

# Draw the graph
plt.figure(figsize=(15, 9))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='blue')

# Draw edges
nx.draw_networkx_edges(G, pos, edgelist=edges, width=2, edge_color='blue')

# Draw and adjust label positions for better readability
# Adjusting the label positions for each station individually
plt.text(0.4, 0, 'Hyde Park Corner', fontsize=12, ha='center', va='center')
plt.text(0.6, 0.7, 'Green Park', fontsize=12, ha='center', va='center')
plt.text(1.6, 0.7, 'Piccadilly Circus', fontsize=12, ha='center', va='center')
plt.text(2.8, 0.5, 'Leicester Square', fontsize=12, ha='center', va='center')
plt.text(3.3, 1, 'Covent Garden', fontsize=12, ha='center', va='center')
plt.text(3.2, 1.5, 'Holborn', fontsize=12, ha='center', va='center')

# Draw edge labels (distances)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black')

# Add a custom line for the legend
plt.plot([], [], color='blue', linewidth=2, label='Piccadilly')
plt.plot([], [], ' ', label='Distances in kilometers') 

# Add a legend for the Piccadilly line
legend = plt.legend(loc='lower right', title='Key', frameon=True, fontsize=12, 
                    title_fontsize=14, labelspacing=1.5, borderaxespad=2, handlelength=5)


# Adjust the legend length and border
frame = legend.get_frame()
frame.set_edgecolor('black')  # Set the edge color
frame.set_linewidth(2)        # Set the line width

# Change the font color of individual labels in the legend
legend.get_texts()[0].set_color("blue")  # Change the color of the first label

# Add a title
plt.title('Piccadilly Line (Partial)')

# Show the plot
plt.show()
```
